```python
#This module helps to clear the output everytime a new board is printed
from IPython.display import clear_output

def display_board(board):
    '''
    Simply displays the playing board
    '''
    clear_output()
    print("---------")
    print(board[1]+" | "+board[2]+" | "+board[3])
    print("----------")
    print(board[4]+" | "+board[5]+" | "+board[6])
    print("----------")
    print(board[7]+" | "+board[8]+" | "+board[9])
    print("---------")
```

```python
def user_input(board):
    '''
    Takes the user input of position to put the X and O and returns back the position
    '''
    check=True
    
    while check:
        position=input("Please select the positon: ")
        if position in board:
            return position
        else :
            print("Please select a Valid position")
```

```python
def placing_input(players_input,value):
    '''
    Places the X/O on the board and returns the updated board
    '''
    location=board.index(value)
    board[location]=players_input    
    return board
```

```python
def check_board(board):
    '''
    Checks if any player has won or the game has tied,
    Returns 1 if Player X has won,
    Returns 2 if Player O has won and
    Returns 3 if the game has tied
    '''
    
    if board[1]=='X' and board[2]=='X' and board[3]=='X':
        return 1
    elif board[4]=='X' and board[5]=='X' and board[6]=='X':
        return 1
    elif board[7]=='X' and board[8]=='X' and board[9]=='X':
        return 1
    elif board[1]=='X' and board[4]=='X' and board[7]=='X':
        return 1
    elif board[2]=='X' and board[5]=='X' and board[8]=='X':
        return 1
    elif board[3]=='X' and board[6]=='X' and board[9]=='X':
        return 1
    elif board[1]=='X' and board[5]=='X' and board[9]=='X':
        return 1
    elif board[3]=='X' and board[5]=='X' and board[7]=='X':
        return 1
    elif board[1]=='O' and board[2]=='O' and board[3]=='O':
        return 2
    elif board[4]=='O' and board[5]=='O' and board[6]=='O':
        return 2
    elif board[7]=='O' and board[8]=='O' and board[9]=='O':
        return 2
    elif board[1]=='O' and board[4]=='O' and board[7]=='O':
        return 2
    elif board[2]=='O' and board[5]=='O' and board[8]=='O':
        return 2
    elif board[3]=='O' and board[6]=='O' and board[9]=='O':
        return 2
    elif board[1]=='O' and board[5]=='O' and board[9]=='O':
        return 2
    elif board[3]=='O' and board[5]=='O' and board[7]=='O':
        return 2
    else:
        return 3
```

```python
def play_again():
    '''
    Play_again checks if the user wants to play again and returns True if yes and False if No
    '''

    check=True
    users_choice=''
    while check:
            users_choice=(input("Do you play again ?? (Y/N)")).upper()
            if users_choice not in ['Y','N']:
                print("Please select a valid option")
            elif users_choice=='Y':
                return True
                check=False
            else:
                return False
                check=False 
```

```python
'''
Initially a board(list) is created,
$ is added as the first element so that the position number display and the index match

'''

game_on=True
position=1

while game_on:
    board=['$','1','2','3','4','5','6','7','8','9']
    display_board(board)

    check=True
    check1=True
    chances=9
    
    while check:
        user_xo_choice=str(input("Please select (X/O):: ")).upper()
    
        if user_xo_choice not in ['X','O']:
            print("Please enter a Valid choice")
        else:
            check=False
        
    while check1:
        
        #taking position from user to input X or O
        position=user_input(board)
        
        #placing the users X or O on the corresponding position on board
        board=placing_input(user_xo_choice,position)
        chances=chances-1
        
        display_board(board)
        
        #switching player turn
        if user_xo_choice=='X':
            user_xo_choice='O'
        elif user_xo_choice=='O':
            user_xo_choice='X'
        
        #will contain 1,2 or 3
        check_board_return=check_board(board)
        
        if check_board_return==1:
            print("Player X wins")
            check1=False
            game_on=play_again()
                        
        elif check_board_return==2:
            print("Player O wins")
            check1=False
            game_on=play_again()
        
        elif chances==0:
            print("The Game is Tied")
            check1=False
            game_on=play_again()

print("Thank You for Playing")
```
